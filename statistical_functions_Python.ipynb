Шаг 2: Выполнение SQL-запросов с использованием Pandas
Запрос 1: Отбор 21-го, 37-го и 47-го элемента вариационного ряда
В SQL вы использовали сортировку с фильтрацией. В Pandas это можно сделать так:

```python
# Фильтрация и сортировка данных по заработной плате, исключая два региона
filtered_df = df[~df['region_name'].isin(['Томская область', 'Самарская область'])].sort_values(by='salary')


# Получение 21-й, 37-й и 47-й строки (индексация с 0, поэтому берем строки 20, 36 и 46)
x_21 = filtered_df.iloc[20]
x_37 = filtered_df.iloc[36]
x_40 = filtered_df.iloc[40]

# Вывод результатов
print(f"X(21): {x_21['region_name']}, {x_21['salary']}")
print(f"X(37): {x_37['region_name']}, {x_37['salary']}")
print(f"X(40): {x_40['region_name']}, {x_40['salary']}")    # неверно почем то "Астраханская область"	32166 надо
filtered_df
```

Шаг 3: Вычисление выборочного среднего
Для вычисления среднего заработка:

```python
# Выборочное среднее
average_salary = filtered_df['salary'].mean()
print(f"Выборочное среднее: {round(average_salary, 2)}")
```

Шаг 4: Определение выборочной медианы
Для нахождения медианы используйте метод median():

```python
# Выборочная медиана
median_salary = filtered_df['salary'].median()
print(f"Выборочная медиана: {median_salary}")
```

Шаг 5: Определение выборочной дисперсии
Для нахождения дисперсии:


```python
# Выборочная дисперсия
variance_salary = filtered_df['salary'].var()
print(f"Выборочная дисперсия: {round(variance_salary, 2)}")
```

Шаг 6: Подсчет количества строк
Для подсчета количества строк:

```python
# Подсчет количества строк
count_salary = filtered_df['salary'].count()
print(f"Количество строк: {count_salary}")
```
